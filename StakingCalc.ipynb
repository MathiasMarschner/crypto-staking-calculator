# Staking vs. Hodling

```python
import math
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#Steuersatz
s = 0.42
```

```python
def time_years(start_price, end_price, taxrate, apy):
    return np.log( (1 - taxrate * (start_price / end_price)) / (1 - taxrate)) / np.log(1 + apy)

def time_days(start_price, end_price, taxrate, apy):
    return time_years(start_price, end_price, taxrate, apy)*365
```

```python
# Beispiel aus Forum: 
# https://forum.blocktrainer.de/t/mathematische-formel-inkl-beispiel-fuer-wann-und-wie-lohnt-sich-staking-trotz-steuern/16615
time_years(2.0,4.0,0.4,0.05)
```

```python
time_days(2.0,10.0,0.43,1.2)
```

```python
start_price = 1.0
apy = 3


end_price = np.linspace(2.0,100.0,1000)
plt.plot(end_price, time_years(start_price, end_price, s, apy))
plt.ylabel('years')
plt.xlabel('end price in multiples of start price')
```

```python
plt.plot(end_price, time_days(start_price, end_price, s, apy))
plt.ylabel('days')
plt.xlabel('end price in multiples of start price')
```

```python
start_price = 1.0
end_price = 10.0
apy = np.arange(50,500,10)
plt.plot(apy, time_days(start_price, end_price, s, apy/100.0))
plt.ylabel('years')
plt.xlabel('apy')
plt.title(f"{end_price/start_price}x")
```

```python
start_price = 1.0
end_price = 3.0
apy = np.arange(50,500,10)
plt.plot(apy, time_days(start_price, end_price, s, apy/100.0))
plt.ylabel('years')
plt.xlabel('apy')
plt.title(f"{end_price/start_price}x")
```
